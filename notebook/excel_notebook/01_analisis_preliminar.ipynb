Revisamos la informacion de cada DF para verificar que las columnas son las mismas, el tipo de dato de cada columna y si hay o no valores nulos (en este caso no los hay)

Unificaremos las distintas hojas del archivo Excel en un único DF, cada hoja se corresponde con años consecutivos, de 2012 a 2014. Mediante la función pd.concat(), ya que todas presentan la misma estructura de variables.

```python
df= pd.concat([df_2012, df_2013, df_2014], ignore_index=True)
```

```python
df.sample(2)
```

Eliminamos la columna "Unnamed: 0" dado que no nos aporta valos al análisis

```python
df=df.drop(columns="Unnamed: 0")
```

```python
df.describe().T
```

```python
for col in df.select_dtypes(include="O").columns:
    df[col] = df[col].str.lower()
```

```python
df.sample()
```

Extraemos el año en una nueva columna, con el fin de facilitar los análisis temporales y las comparaciones interanuales.

```python
df["año"] = df["Dt_Customer"].dt.year
```

```python
df.sample()
```

```python
df["Kidhome"].value_counts()
```

```python
df["Teenhome"].value_counts()
```

Se normalizan los nombres de las variables a minúscula para garantizar consistencia con el DF "bank-additional", respetando la columna "ID" clave para la identificación para la posterior unión.

```python

df.columns= [col.lower() if col != "ID" else col
for col in df.columns]
```

```python
df.sample(2)
```

```python
df.duplicated().sum()
```

No hay duplicados en este DataFrame

```python
df.groupby("kidhome")["income"].mean()
```

```python
df.groupby("teenhome")["income"].mean()
```

En un primer análisis sencillo, vemos que a medida que sube el numero de hijos en casa, sube el nivel medio de ingresos. Ahora queremos comprobar si las personas que más utilizan la aplicacion del banco son aquellas que más dinero tienen.

```python
df.groupby("numwebvisitsmonth")["income"].mean().sort_index(ascending=False)
```

Vemos que la variable "veces que entra en la web" no esta relacionada con el nivel salarial que los clientes tienen, ya que vemos que para el valor "4" la media salarial es de  94620.29€ y para el valor "32" la media salarial es de 94245.93€

Dado que ya hemos revisado de forma indidual los datos de cada DF antes de la unión y que ya hemos hecho los cambios oportunos en el DF unido, procedemos a guardar estos datos en un nuevo archivo csv para proceder a la union con el csv ya trabajado "bank-additional.csv"
