# Limpieza general de los datos

### Ponemos todos los valores de las columnas "object" en minúsculas

```python
for col in df_bank.select_dtypes(include="O").columns:
    df_bank[col] = df_bank[col].str.lower()

df_bank.tail(5)
```

### Cambiamos todas las "," por "." que hay en las columnas categorizadas como "object" y las dejamos en tipo "float"

```python
def comas (df):
    for col in df.select_dtypes(include="O"):

        df[col] = df[col].str.replace(",", ".")
        try:
            df[col] = df[col].astype("float64")
        except:
            pass
comas (df_bank)
df_bank.info()
```

```python
df = df_bank.copy()
```

### Modificamos la columna *date* a tipo fecha

```python
meses = {
    "enero": "01",
    "febrero": "02",
    "marzo": "03",
    "abril": "04",
    "mayo": "05",
    "junio": "06",
    "julio": "07",
    "agosto": "08",
    "septiembre": "09",
    "octubre": "10",
    "noviembre": "11",
    "diciembre": "12"
}

df['date'] = df['date'].str.lower()
df['date'] = df['date'].replace(meses, regex=True)
```

```python
df["date"]= pd.to_datetime(df["date"], format="%d-%m-%Y", errors='coerce')
```
